```python
import pandas as pd

# Load Excel (no header row)
df = pd.read_excel("people.xlsx", header=0)

# Rename useful columns for easier access
df = df.rename(columns={
    df.columns[1]: "lastname",
    df.columns[2]: "firstname",
    df.columns[3]: "email"
})

def search_person(query):
    q = query.lower()

    results = df[
        df["firstname"].astype(str).str.lower().str.contains(q) |
        df["lastname"].astype(str).str.lower().str.contains(q)
    ]

    return results

while True:
    q = input("Search: ").strip()
    if q == "":
        print("No input.\n")
        break
    results = search_person(q)

    if results.empty:
        print("No match.\n")
        break
    else:
        print("\n=== RESULT ===")
        for _, row in results.iterrows():
            print(f"{row['lastname']}\t{row['firstname']}\t{row['email']}")
        print("==============\n")
```
